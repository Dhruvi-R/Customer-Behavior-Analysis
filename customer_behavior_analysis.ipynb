```python
#everything in lowercase and space replace with underscore
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
df=df.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
df.columns
```

```python
#Create Column a age_group
labels=['Young Adult','Adult','Middle-aged','Senior']
df['age_group']=pd.qcut(df['age'],q=4,labels=labels)
```

```python
frequency_mapping={'Fortnightly':14,
                   'Weekly':7,
                   'Annually':365,
                   'Quarterly':90,
                   'Bi-Weekly':14,
                   'Monthly':30,
                   'Every 3 Months':90
                  }
df['purchase_frequency_days']=df['frequency_of_purchases'].map(frequency_mapping)
```

```python
# Q1:What is the total revenue generated by Male VS Female Customers?
q1=pd.read_sql("""SELECT gender as Gender,SUM(purchase_amount) as Revenue 
FROM CUSTOMER
GROUP BY gender""",conn)
q1
```

```python
# Q2: Which customer used a discount but still spent more than the average purchase amount ?
q2=pd.read_sql("""SELECT customer_id,purchase_amount
               FROM CUSTOMER WHERE discount_applied='Yes' AND purchase_amount>=(SELECT AVG(purchase_amount) FROM CUSTOMER)""",conn)
q2
```

```python
# Q3: Which are the top 5 products with the highest average review rating?
q3=pd.read_sql("""SELECT item_purchased,ROUND(AVG(review_rating),2) AS AVG_PRODUCT_RATING 
FROM CUSTOMER
GROUP BY item_purchased
ORDER BY AVG_PRODUCT_RATING DESC
LIMIT 5
""",conn)
q3
```

```python
# Q4: Compare the average purchase amount between Standard and Express Shipping 
q4=pd.read_sql(""" SELECT shipping_type,
                ROUND(AVG(purchase_amount),2) AS purchase_amount
                FROM CUSTOMER
                WHERE shipping_type IN ('Standard','Express')
                GROUP BY shipping_type""",conn)
q4
```

```python
#Q5 Do subscribed customer spend more ?
#Compare average spend and total revenue between subscribers and non subscribers.
q5=pd.read_sql("""SELECT subscription_status AS SUBSCRIPTION_STATUS,
                COUNT(customer_id) AS TOTAL_CUSTOMER,
                AVG(purchase_amount) AS AVERAGE_SPEND,
                SUM(purchase_amount) AS TOTAL_REVENUE
                FROM CUSTOMER
                WHERE subscription_status IN ('Yes','No')
                GROUP BY subscription_status""",conn)
q5
```

```python
#Q6 Which 5 products have the highest percentage of purchase with discount applied ?
q6=pd.read_sql("""SELECT item_purchased,
                ROUND(100*SUM(CASE WHEN discount_applied='Yes' THEN 1 ELSE 0 END)/COUNT(*),2) AS DISCOUNT_RATE
                FROM CUSTOMER 
                GROUP BY item_purchased
                ORDER BY DISCOUNT_RATE DESC
                LIMIT 5 """,conn)
q6
```

```python
# Q7: Segment customer into New, Returning, Loyal based on thier Total- 
# number of previous purchases , and show the count of each segments.
Q7=pd.read_sql("""WITH CUSTOMER_TYPE AS
                (SELECT customer_id,previous_purchases,
                CASE 
                WHEN previous_purchases=1 THEN 'New'
                WHEN previous_purchases BETWEEN 2 AND 10 THEN 'Returning'
                ELSE 'Loyal'
                END AS CUSTOMER_SEGMENTS
                FROM CUSTOMER)
                SELECT CUSTOMER_SEGMENTS,COUNT(*) AS Number_of_Customer
                FROM CUSTOMER_TYPE 
                GROUP BY CUSTOMER_SEGMENTS""",conn)
Q7
```

```python
# Q8: What are the top 3 most purchased products withing each category?
q8 = pd.read_sql("""
WITH item_counts AS (
    SELECT 
        category,
        item_purchased,
        COUNT(customer_id) AS total_orders,
        ROW_NUMBER() OVER (
            PARTITION BY category 
            ORDER BY COUNT(customer_id) DESC
        ) AS item_rank
    FROM CUSTOMER
    GROUP BY category, item_purchased
)
SELECT 
    item_rank,
    category,
    item_purchased,
    total_orders
FROM item_counts
WHERE item_rank <= 3;
""", conn)

q8
```

```python
# Q9 : Are Customer who are repeat buyers (more than 5 previous purchases) also likely to subscribe?
q9=pd.read_sql("""SELECT COUNT(customer_id),subscription_status
                FROM CUSTOMER 
                WHERE previous_purchases>=5 
                GROUP BY subscription_status
                """,conn)


q9
```

```python
# Q10: What is the revenue contribution of each age group ?
q10=pd.read_sql("""SELECT age_group,SUM(purchase_amount) AS REVENUE
                FROM CUSTOMER
                GROUP BY age_group	
                """,conn)
q10
```
